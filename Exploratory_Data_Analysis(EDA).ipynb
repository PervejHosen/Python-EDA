# **What is an Outlier**

A data point which is significantly far away from other data points
* IQR (Inter Quartile Range)
* IQR = Q3 - Q1
* Lower Boundary = Q1 - (1.5 * IQR)
* Upper Boundary = Q3 + (1.5 * IQR)

```python
data = pd.read_csv('/content/drive/MyDrive/Python/Datasets/EDA/house_prices.csv')
```

```python
data.head()
```

```python
data['SalePrice'].describe()
```

```python
sns.boxplot(x=data['SalePrice'])
```

```python
data.shape
```

```python
# Let's calculate first quartile,third_quartile and inter_quartile range
first_quartile = data['SalePrice'].quantile(.25)
third_quartile = data['SalePrice'].quantile(.75)
IQR = third_quartile - first_quartile
```

```python
new_boundary = third_quartile + 3 * IQR
```

```python
# dropped the outliers data
new_data = data.drop(data[data['SalePrice']>new_boundary].index, axis = 0, inplace=False)
```

```python
# 12 rows are dropped
new_data.shape
```

```python
sns.boxplot(x=new_data['SalePrice'])
```

1.2 Outliers detection using IQR

```python
# height of a persons
raw ={
    'name': ['mohan','maria','deepak','kunal','piyush','avinash','lisa','smita','tanu', 'khusboo','nishant','johnson','donald','rakesh','pritvi','roy','ashish','abhishek','jassi'],
    'height': [1.2, 2.3, 4.9, 5.1, 5.2, 5.4, 5.5, 5.5, 5.6, 5.6, 5.8, 5.9, 6, 6.1, 6.2, 6.5, 7.1, 14.5, 23.2,]
    }
df =pd.DataFrame(raw)
```

```python
df
```

```python
df.describe()
```

```python
sns.boxplot(x=df['height'])
```

```python
# Calculating the Q1 ,Q3
Q1 = df.height.quantile(.25)
Q3 = df.height.quantile(.75)
Q1 , Q3
```

```python
 # Calculating IQR
IQR = Q3 - Q1
IQR
```

```python
# Calculating IQR
lower_limit = Q1 - 1.5 * (IQR)
upper_limit = Q3 + 1.5 * (IQR)
lower_limit,upper_limit
```

```python
#outliers
df[(df.height<lower_limit)|(df.height>upper_limit)]
```

```python
# data with no outliers
df_no_outliers = df[(df.height>lower_limit) & (df.height < upper_limit) ]
```

```python
sns.boxplot(x=df['height'])
```

```python
df_no_outliers
```

1.3 Outlier detection and removal using Z score

Z score indicates how many standard deviation away a datapoint is.
* For example in our case mean is 66.37 and the standard deviation is 3.84
If a value of data point is 77.91, then Z score for that is 3 because it is 3 standard
deviation away (77.91 = 66.37 + 3 * 3.84)
Score is refer to data point below


1.4 z = (x – ) /

```python
df['zscore'] = (df.Height - df.Height.mean())/df.Height.std()
```

```python
df.head()
```

```python
#outliers
df[(df['zscore']>3) | (df['zscore']<-3)]
```

```python
#removing outliers as per Z score
df_no_outliers = df[(df.zscore <3) & (df.zscore > -3)]
df_no_outliers.head()
```

```python
#checkig the number of ouliers
df.shape[0] - df_no_outliers.shape[0]
```

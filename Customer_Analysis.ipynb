# **Customer Analytical Data Report**

## Overview

When analyzing customer analytical data, we encounter two primary types of data: **Market Basket Data** and **Sales Data**. Each type provides unique insights into customer behavior and product performance.

## 1. Market Basket Data

Market basket data relates to the orders made by customers. It helps us understand the purchasing patterns and relationships among the items they buy. This type of data answers questions like:

### Key Insights

- **Customer Patterns:**
  - **Definition:** Identifies the combinations of products that customers tend to purchase together.
  - **Example:** Customers often buy toilet paper, detergent, and milk in one trip, even though these items aren’t directly related.
  
- **Common Bundles:**
  - **Definition:** Identifies which products are typically bundled together in purchases.
  - **Example:** Fruits, vegetables, and sauces are often bought together by customers planning to cook meals.

## 2. Sales Data

Sales data focuses on the performance of individual items, known as SKUs (Stock Keeping Units). This type of data answers questions related to product sales, revenue, and profitability.

### Key Insights

- **Top Performers:**
  - **Definition:** Identifies which items are selling the most.
  - **Example:** Determining the highest revenue-generating products.
  
- **Profitability:**
  - **Definition:** Identifies which items provide the highest profit margins.
  
- **Customer Attraction:**
  - **Definition:** Identifies which items draw the most customers into the store.
  - **Example:** Determining items that generate significant footfall.
  
- **Slow Movers:**
  - **Definition:** Identifies items that are not selling well and might need a discount or markdown.

## Hierarchy of Products

To understand how products are organized, consider the following hierarchical structure:

1. **Department:** Broad category (e.g., Grocery)
2. **Product Families:** Subdivisions within departments (e.g., Dairy)
3. **Subfamilies:** More specific groups (e.g., Milk products)
4. **Sections:** Further division (e.g., Flavored Milk)
5. **SKUs:** The most specific level, individual items (e.g., 1-liter Chocolate Milk)

## Practical Applications

### Market Basket Analysis

- **Store Layout:**
  - **Strategy:** Place frequently bought-together items near each other.
  - **Benefit:** Makes shopping convenient and potentially increases sales.
  
- **Promotions:**
  - **Strategy:** Create deals or discounts for common bundles.
  - **Example:** Offering a discount when buying both toilet paper and detergent.

### Sales Data Analysis

- **Stock Decisions:**
  - **Strategy:** Determine which items need to be restocked frequently.
  - **Benefit:** Ensures availability based on sales performance.
  
- **Marketing Strategies:**
  - **Strategy:** Focus on promoting high-profit or high-revenue items.
  
- **Inventory Management:**
  - **Strategy:** Identify slow-moving items that may need markdowns.
  - **Benefit:** Helps clear stock and optimize inventory.

## Summary

- **Market Basket Data:** Provides insights into customer purchasing patterns by analyzing what items are bought together.
- **Sales Data:** Focuses on the performance of individual items, helping determine which products sell well, generate revenue, and attract customers.

```python
'''

Market Basket Analysis
What is Market Basket Analysis?
Market Basket Analysis is a data mining technique used to uncover relationships between items in a dataset, typically in a retail setting. It helps identify items that frequently co-occur in transactions. This analysis is commonly used to understand customer purchasing behavior, optimize product placement, and create effective marketing strategies.

How Does Market Basket Analysis Work?
Market Basket Analysis examines transaction data to find associations between items. The primary method used is the Apriori algorithm, which identifies itemsets that frequently appear together in transactions.

Simple Example
Imagine you own a grocery store and have transaction data from your customers. Here’s a simplified example:

Transaction 1: Bread, Butter
Transaction 2: Bread, Butter, Milk
Transaction 3: Bread, Milk
Transaction 4: Butter, Milk
By analyzing this data, you might find that:

Customers who buy bread often buy butter.
Customers who buy butter might also buy milk.
From these insights, you can create promotions, such as a discount on butter if a customer buys bread, or place these items closer together in the store.

Key Concepts in Market Basket Analysis
Support: The proportion of transactions that contain a specific itemset.

Example: If 3 out of 4 transactions include bread, the support for bread is 75%.
Confidence: The likelihood that a transaction containing item A also contains item B.

Example: If 3 out of 3 transactions containing bread also contain butter, the confidence of the rule "bread → butter" is 100%.
Lift: The ratio of the observed support to that expected if A and B were independent.

Example: If bread and butter are bought together more often than would be expected if they were independent, the lift is greater than 1.
Machine Learning and Statistical Techniques for Market Basket Analysis
Apriori Algorithm:

How it works: Iteratively identifies frequent itemsets and builds association rules from these itemsets.
Use case: Discovering frequent itemsets in a large dataset of transactions.
FP-Growth Algorithm:

How it works: Uses a compressed representation of the transaction database called an FP-tree and extracts frequent itemsets directly.
Use case: More efficient than Apriori for large datasets.
Association Rule Learning:

How it works: Generates rules that describe how often items are purchased together.
Use case: Creating actionable insights for cross-selling and promotions.
Eclat Algorithm:

How it works: Uses a depth-first search approach to find frequent itemsets.
Use case: Faster in certain scenarios compared to Apriori, especially with dense datasets.
Statistical Techniques:

Chi-Square Test: Tests the independence of two items.
Lift Analysis: Measures the strength of an association rule.
Simple Summary
Market Basket Analysis: A technique to identify associations between items in transaction data.
Example: Discovering that bread and butter are frequently bought together.
Key Metrics: Support, confidence, and lift.
Techniques: Apriori algorithm, FP-Growth algorithm, Association Rule Learning, Eclat algorithm, and statistical methods like Chi-Square and lift analysis.

'''
```

```python
'''
RFM Analysis
What is RFM Analysis?
RFM Analysis is a marketing technique used to evaluate and segment customers based on three key metrics: Recency, Frequency, and Monetary value. This method helps businesses understand their customers' behaviors and value, allowing them to tailor marketing strategies and improve customer retention.

Key Metrics in RFM Analysis
Recency (R): How recently a customer made a purchase.
Frequency (F): How often a customer makes a purchase.
Monetary Value (M): How much money a customer spends on purchases.
How Does RFM Analysis Work?
RFM analysis involves scoring customers on each of the three metrics, usually on a scale of 1 to 5, where 5 is the highest value. These scores are then combined to create an overall RFM score for each customer.

Example of RFM Analysis
Imagine you run an online clothing store. Here’s how you might use RFM analysis:

Data Collection: Gather transaction data for your customers, including purchase dates, number of purchases, and total spending.

Scoring:

Recency: Assign scores based on how recently customers made their last purchase. For example, customers who purchased within the last month might get a score of 5, while those who purchased over a year ago get a score of 1.
Frequency: Assign scores based on how many purchases customers made in the past year. Customers with more purchases get higher scores.
Monetary Value: Assign scores based on the total amount spent by customers in the past year. Higher spenders get higher scores.
Combining Scores: Combine the R, F, and M scores to get an overall RFM score for each customer.

Example Calculation:

Customer A:

Recency: Purchased 1 month ago (Score: 5)
Frequency: Made 10 purchases in the past year (Score: 5)
Monetary Value: Spent $1000 in the past year (Score: 5)
RFM Score: 555
Customer B:

Recency: Purchased 6 months ago (Score: 3)
Frequency: Made 4 purchases in the past year (Score: 3)
Monetary Value: Spent $200 in the past year (Score: 2)
RFM Score: 332
Using RFM Analysis
Once customers are scored, you can segment them based on their RFM scores and tailor your marketing strategies accordingly.

Top Customers (High R, F, M Scores): These customers are highly engaged and valuable. Reward them with exclusive offers and personalized communications.
At-Risk Customers (Low R, High F and M Scores): These customers used to purchase frequently and spend a lot but haven't bought anything recently. Re-engage them with special incentives or reminders.
New Customers (High R, Low F and M Scores): These customers have made recent purchases but haven’t yet shown loyalty. Encourage them with follow-up promotions and excellent service.
Low-Value Customers (Low R, F, M Scores): These customers purchase infrequently and spend little. Use cost-effective marketing strategies to encourage occasional purchases.
Machine Learning and Statistical Techniques for RFM Analysis
K-Means Clustering:

How it works: Groups customers into clusters based on their RFM scores.
Example: Segmenting customers into distinct groups such as "Loyal Customers," "At-Risk Customers," and "New Customers."
Decision Trees:

How it works: Classifies customers based on their RFM scores into different segments.
Example: Identifying the key factors (R, F, or M) that determine whether a customer belongs to a high-value segment.
Logistic Regression:

How it works: Predicts the likelihood of a customer responding to a marketing campaign based on their RFM scores.
Example: Estimating which customer segments are more likely to respond to a discount offer.
Simple Summary
RFM Analysis: A technique to evaluate and segment customers based on Recency, Frequency, and Monetary value.
Importance: Helps tailor marketing strategies, improve customer retention, and identify high-value customers.
Example Calculation: Assign scores to customers based on how recently they purchased, how often they purchase, and how much they spend.
Machine Learning and Statistical Techniques:

K-Means Clustering: Groups customers into clusters based on RFM scores.
Decision Trees: Classifies customers into different segments using RFM scores.
Logistic Regression: Predicts customer behavior based on RFM scores.

'''
```

```python
'''
There are several different trade area modeling techniques used in retail and economic analysis. Here are some common ones:

Gravity Models:
Gravity models are based on the principle of gravity, where the attraction between two objects is directly proportional to their masses and inversely proportional to the square of the distance between them. In trade area analysis, this model predicts the flow of goods or people between two locations based on their populations and the distance between them.

The Huff Model:
The Huff model estimates the probability that a customer located at a particular point will choose a specific store or location for shopping. It takes into account factors such as store attractiveness (like size, variety, and reputation) and distance to competing stores. This model helps retailers understand how the presence of competitors affects their market share.

The Reilly's Law of Retail Gravitation:
Reilly's Law of Retail Gravitation is similar to gravity models but incorporates factors such as market size and attractiveness. It predicts the trade area boundaries between competing cities or regions based on their relative size and attractiveness. The larger and more attractive city is expected to draw customers from the smaller and less attractive one.

Analog Approach:
The analog approach involves identifying similar trade areas to the one being analyzed and using their characteristics as a basis for estimation. This method relies on finding comparable areas with similar demographics, economic factors, and competitive landscapes to make predictions about a new trade area.

Agent-Based Models:
Agent-based models simulate the behavior of individual agents (such as consumers or businesses) and their interactions within a trade area. These models can capture complex dynamics and feedback loops, allowing researchers to explore how changes in one aspect (like pricing or advertising) might impact the overall trade area.

Market Penetration Models:
Market penetration models estimate the potential market share a business can capture within a specific trade area. These models consider factors like population demographics, consumer behavior, competition, and the business's own strengths and weaknesses to predict its performance in the market.

Geographic Information Systems (GIS) Analysis:
GIS analysis combines spatial data with other types of information (like demographic, economic, or transportation data) to create visual representations of trade areas. GIS tools allow analysts to map trade areas, identify hotspots or underserved areas, and analyze spatial relationships between different factors influencing trade.


'''
```

```python
'''
Trade areas can be categorized into different types based on various criteria such as geographical scope, market characteristics, or the nature of economic activities within them. Here are some common types of trade areas:

Primary Trade Area (PTA):
The primary trade area refers to the immediate geographical area surrounding a business or retail location from which it draws the majority of its customers. This area typically accounts for the highest proportion of sales and includes nearby neighborhoods or communities.

Secondary Trade Area (STA):
The secondary trade area extends beyond the primary trade area and includes additional areas from which the business attracts customers, albeit to a lesser extent. Customers in the secondary trade area may be willing to travel slightly farther or less frequently to patronize the business.

Tertiary Trade Area (TTA):
The tertiary trade area represents the outermost region from which the business still attracts some customers, although they are typically less frequent or more distant than those in the primary and secondary trade areas. Tertiary trade areas can extend over larger geographical areas and may encompass multiple neighborhoods or even neighboring towns.

Rural Trade Area:
Rural trade areas are characterized by lower population densities and fewer retail options compared to urban or suburban areas. Businesses in rural trade areas may serve a wider geographic area but cater to a smaller customer base.

Urban Trade Area:
Urban trade areas are located within densely populated urban centers and typically have a high concentration of businesses, commercial activities, and consumer demand. These areas often feature a diverse mix of retail, dining, entertainment, and service establishments.

Suburban Trade Area:
Suburban trade areas are situated in the outskirts of urban centers and are characterized by lower population densities compared to urban areas but higher than rural areas. They often cater to a mix of residential and commercial activities, with a focus on shopping centers, strip malls, and residential neighborhoods.

Specialized Trade Area:
Specialized trade areas focus on specific market niches or industries, such as technology parks, industrial estates, or tourist destinations. These areas may have unique characteristics and attract businesses and consumers with specialized needs or interests.

International Trade Area:
International trade areas encompass regions or zones where cross-border trade and economic activities take place between different countries or regions. These areas can include ports, free trade zones, or international airports and play a crucial role in global commerce and supply chains.

'''
```

```python
'''
Trade area analysis helps businesses understand the geographic area from which they draw customers. Here are explanations of three commonly used methods: drive-time analysis, radial studies, and gravity models, along with an additional method called the Thiessen polygon method.

1. Drive-Time Analysis
Explanation:
Drive-time analysis calculates the area that can be reached within a specific driving time from a particular location. This method considers road networks, traffic conditions, and speed limits to determine how far customers can travel within a given time frame.

Example:
Imagine you are opening a new coffee shop and want to know your potential customer base. Using drive-time analysis, you determine the area that customers can reach within a 10-minute drive. This helps you understand which neighborhoods and businesses fall within this zone, enabling you to tailor your marketing efforts to those areas.

2. Radial Studies
Explanation:
Radial studies, also known as ring studies, analyze trade areas by drawing circles (rings) of specific radii around a business location. These rings typically represent different levels of market penetration, such as primary, secondary, and tertiary trade areas.

Example:
For a new retail store, you might draw rings at 1 mile, 3 miles, and 5 miles around the store location. The 1-mile ring represents the primary trade area where most customers are expected to come from, the 3-mile ring is the secondary trade area, and the 5-mile ring is the tertiary trade area. This visual helps you see which areas contribute most to your customer base and plan your advertising and outreach accordingly.

3. Gravity Models
Explanation:
Gravity models predict the likelihood of customers visiting a particular store based on the store's attractiveness (such as size, variety, and reputation) and the distance from the customer's location. The model is based on the principle that larger and closer stores attract more customers.

Example:
Suppose you have two shopping malls in a city: Mall A and Mall B. Mall A is larger and has more stores, but it is farther from a specific residential area. Mall B is smaller but closer to the same area. A gravity model would help predict which mall residents are more likely to visit based on the attractiveness of the malls and their distance from the residential area. This can help mall owners understand their competitive position and customer draw.

4. Thiessen Polygon Method
Explanation:
The Thiessen polygon method, also known as Voronoi polygons, divides a geographic area into regions based on the proximity to a set of points. Each polygon contains all the locations that are closest to a particular point, such as store locations.

Example:
If you have three grocery stores in a city, the Thiessen polygon method would create three polygons, each representing the area closest to one of the stores. This method helps identify which areas are predominantly served by each store and can reveal regions that might be underserved or highly competitive. For instance, if a new residential development falls within the polygon of one store, that store can focus its marketing efforts on this new potential customer base.

Summary
Each trade area analysis method has its strengths and applications:

Drive-Time Analysis is great for understanding accessibility based on real travel times.
Radial Studies provide a simple and clear visualization of market penetration based on distance.
Gravity Models offer insights into customer behavior by considering both attractiveness and distance.
Thiessen Polygon Method helps in visualizing and managing service areas and competitive landscapes.
By using these methods, businesses can make informed decisions about location planning, marketing strategies, and customer outreach.

'''
```

```python
scaler = MinMaxScaler()

# Select the columns to scale (excluding the 'customer community' column)
columns_to_scale = attractivness.columns[1:]

# Fit and transform the data
attractivness[columns_to_scale] = scaler.fit_transform(attractivness[columns_to_scale])
```

```python
attractivness
```

```python
# Get unique stores and communities
stores = attractivness['customer community'].unique()
communities = distances['customer community'].unique()

# Create keys for each community-store pair
keys = [(x, y) for x in communities for y in stores]
```

```python
# Scale the attractiveness data
scaled = attractivness.copy()
for column in ['Size', 'Parking spaces', 'Highways', 'Traffic', 'Accessibility', 'Design']:
    scaled[column] = (scaled[column] - scaled[column].min()) / (scaled[column].max() - scaled[column].min())

# Calculate overall attractiveness score for each store
scaled['attractivness'] = (scaled['Size'] + scaled['Parking spaces'] + scaled['Highways'] +
                            scaled['Traffic'] + scaled['Accessibility'] + scaled['Design'])
```

```python
# Initialize an empty dictionary for the numerator
numerator = {}

for community, store in keys:
    attractiveness_score = scaled.loc[scaled['customer community'] == store, 'attractivness'].values[0]
    distance_value = distances.loc[distances['customer community'] == community, store].values[0]
    numerator[(community, store)] = attractiveness_score / distance_value**2
```

```python
# Initialize an empty dictionary for Pijs
Pijs = {}

for community, store in keys:
    sum_numerator = sum([value for (comm, st), value in numerator.items() if comm == community])
    Pijs[(community, store)] = numerator[(community, store)] / sum_numerator
```

```python
# Initialize an empty dictionary for expected sales per key
expected_sales_per_key = {}

for community, store in keys:
    total_expenditure = census.loc[census['customer community'] == community, 'total'].values[0]
    expected_sales_per_key[(community, store)] = Pijs[(community, store)] * total_expenditure

# Initialize an empty dictionary for expected sales per store
expected_sales_per_store = {}

for store in stores:
    expected_sales_per_store[store] = sum([value for (comm, st), value in expected_sales_per_key.items() if st == store])

expected_sales_per_store
```

```python
'''
 change the data type of the “PurchaseDate” from object to “datetime” type

'''

from datetime import datetime
df['PurchaseDate'] = pd.to_datetime(df['PurchaseDate'])


```

```python
'''
Recency value calculation
The “Recency” for each customer is  calculated by counting the number of days between the current date and the last purchase date.

'''

df['Recency'] = (datetime.now() - df['PurchaseDate']).dt.days
```

```python
'''
Frequency value calculation
The “Frequency” is then calculated by grouping the data by ‘CustomerID’ and calculating the number of orders for each customer

'''


frequency_data = df.groupby('CustomerID')['OrderID'].count().reset_index()
frequency_data.rename(columns= {'OrderID': 'Frequency'}, inplace = True)
df = df.merge(frequency_data, on = 'CustomerID', how = 'left')
```

```python
'''
Monetary value calculation
compute the “Monetary Value” by grouping the data by ‘CustomerID’ and summing the transaction amounts for each customer.

'''


monetary_data = df.groupby('CustomerID')['TransactionAmount'].sum().reset_index()
monetary_data.rename(columns={'TransactionAmount': 'MonetaryValue'}, inplace = True)
df = df.merge(monetary_data, on = 'CustomerID', how = 'left')
```

```python
df
```

```python
'''
Calculating RFM Scores
“Recency” scores are higher for customers with more recent interactions,
“Frequency” scores are higher for customers with higher purchase frequency;
and “Monetary Value” scores are higher for customers who spend more money

'''

recency_scores = [5, 4, 3, 2, 1] # Scoring Criteria
frequency_scores = [1, 2, 3, 4, 5]
monetary_scores = [1, 2, 3, 4, 5]

df['RecencyScore'] = pd.cut(df['Recency'], bins = 5, labels= recency_scores)
df['FrequencyScore'] = pd.cut(df['Frequency'], bins = 5, labels= frequency_scores)
df['MonetaryScore'] = pd.cut(df['MonetaryValue'], bins= 5, labels= monetary_scores)

'''
 RecencyScore, FrequencyScore, and MonetaryScore columns give a quantitative
 picture of each customer’s RFM profile,
  which may be utilized for segmentation and further analysis in marketing campaigns.
'''

```
